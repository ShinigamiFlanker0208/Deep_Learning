# Pure Functions (No Hidden Surprises)

JAX's magic tools—like jax.jit (which compiles your code to make it blindingly fast) and jax.grad (which calculates your gradients)—only work on pure functions.

A pure function means:

It doesn't rely on global variables.

It doesn't print things or write to files (no side effects).

If you give it the same inputs, it always returns the exact same outputs.

```python
y=2
def impure_function(x):
    print("Side effect!")
    return x + y

print(impure_function(1))
```

```python
# JAX works on these fundamentals

def pure_function(x,y):
    return x + y
x=2
y=2
print(pure_function(x,y))
```

Rule 2: No Side Effects (No touching the outside)
While the function is running, it is not allowed to change the state of the rest of your program. It cannot modify global lists, save files, update database records, or even rely on print() statements to track logic.
