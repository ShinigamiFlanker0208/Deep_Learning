# Neural Network from scratch

```python
import numpy as np
```

```python
def init_params():
    # W1 = 10 neurons, 784 inputs each
    W1 = np.random.randn(10, 784) * np.sqrt(2. / 784)
    b1 = np.zeros((10, 1))

    # W2 = 10 outputs (classes 0-9), 10 inputs from hidden layer
    W2 = np.random.randn(10, 10) * np.sqrt(2. / 10)
    b2 = np.zeros((10, 1))

    return W1, b1, W2, b2
```

```python
# Forward Propagation

def forward_prop(W1,b1,W2,b2,X):
    Z1=np.dot(X,W1)+b1
    A1=np.maximum(0,Z1)
    Z2=np.dot(A1,W2)+b2
    A2=np.exp(Z2)/np.sum(np.exp(Z2), axis=0)
    return Z1,A1,Z2,A2
```

```python
# BackPropogation
def back_prop(Z1,A1,Z2,A2,X,Y):
    m=Y.size
    one_hot_Y=np.zeros((10,m))
    one_hot_Y[Y,np.arange(m)]=1 #Encode Labels

    dZ2=A2-one_hot_Y
    dW2=np.dot(A1,dZ2.T)/m
    db2=np.sum(dZ2, axis=1, keepdims=True)/m
    dZ1=np.dot(dZ2,W2.T)
    dW1=np.dot(X,dZ1.T)/m
    db1=np.sum(dZ1, axis=1, keepdims=True)/m
    return dW1,db1,dW2,db2
```
